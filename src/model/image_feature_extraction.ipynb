```python
df_raw = pd.read_json("/kaggle/input/data-products-40k/list_products_40000.json", lines = True)
df = df_raw.sample(n = df_raw.shape[0], random_state = 1)
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.resnet50 import preprocess_input
```

```python
def extract_image_to_vector():
  global df, df_raw
  features = []
  indexs = df.index
  i = 0
  for index in indexs:
    try:
      image = Image.open("/kaggle/input/data-products-40k/images_40000/images_40000/{index_image}.png".format(index_image = index))
      image = image.convert("RGB")
      image = image.resize((256, 256))
      image = np.asarray(image)
      image = preprocess_input(np.expand_dims(image.copy(), axis=0))

      resnet_feature = my_new_model.predict(image)
      resnet_feature_np = np.array(resnet_feature)
      features.append(resnet_feature_np.flatten())
      i += 1
    except Exception as e:
      print("The error is: ",e)
      print(index,image, df.at[index, 'url_thumbnail'])
      df.drop(index, inplace=True)
      df_raw.drop(index, inplace=True)
      i += 1

  return features
```

```python
resnet = ResNet50(include_top=False, pooling='avg', weights='imagenet')
my_new_model = Sequential()
my_new_model.add(resnet)
my_new_model.layers[0].trainable = False
```

```python
ImageFile.LOAD_TRUNCATED_IMAGES=True
vectorized_product_images = extract_image_to_vector()
```
